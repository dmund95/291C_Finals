```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
```

```python
s_arr = np.linspace(0, 1, 10001)
num_s = len(s_arr)
s_arr_idx = np.arange(0,num_s)
assert(len(s_arr) == len(s_arr_idx))

a_arr = np.linspace(0.01, 0.2, 20)
num_a = len(a_arr)
a_arr_idx = np.arange(0,num_a)
assert(len(a_arr) == len(a_arr_idx))


r_arr = np.array([-1]*num_s)
r_arr[0] = 1000
r_arr_idx = np.arange(0,num_s)
assert(len(r_arr) == len(r_arr_idx))

eps_arr = np.linspace(0.01,0.1, 10)

q_arr = np.zeros((num_s, num_a))
```

```python
def pick_action(q_arr, s_idx, eps):
    if random.uniform(0, 1) < eps:
        return random.choice(a_arr_idx)
    else:
        return np.argmax(q_arr[s_idx,:])
```

```python
def find_nearest(array,value):
    idx = np.searchsorted(array, value, side="left")
    if idx > 0 and (idx == len(array) or math.fabs(value - array[idx-1]) < math.fabs(value - array[idx])):
        return idx-1
    else:
        return idx
    
def simulate_one_step(s, a):
    next_s = s - 4*a*s*s*s
    return find_nearest(s_arr, next_s)
```

```python
def q_learning(gamma, alpha):
    while(1):
#         curr_s_idx = random.choice(s_arr_idx)
        curr_s_idx = 10000
        eps = random.choice(eps_arr)
        
        while (1):
            a_idx = pick_action(q_arr, curr_s_idx, eps)
            if (curr_s_idx == 0):
                q_arr[curr_s_idx, a_idx] = q_arr[curr_s_idx, a_idx] + alpha*(r_arr[curr_s_idx] - q_arr[curr_s_idx, a_idx])
                break
            else:
                next_s_idx = simulate_one_step(s_arr[curr_s_idx], a_arr[a_idx])
                q_arr[curr_s_idx, a_idx] = q_arr[curr_s_idx, a_idx] + alpha*(r_arr[curr_s_idx] + gamma*np.max(q_arr[next_s_idx,:]) - q_arr[curr_s_idx, a_idx])
                if (curr_s_idx == next_s_idx):
#                     print(s_arr[curr_s_idx])
                    break
                curr_s_idx = next_s_idx
                
                
                
```

```python
q_learning(0.99, 0.1)
```

```python
x_arr = []
y_arr = []
for i in range(num_s):
    if (np.min(q_arr[i,:]) != 0):
        x_arr.append(s_arr[i])
        y_arr.append(a_arr[np.argmin(q_arr[i,:])])

plt.plot(x_arr, y_arr, label='q_learning_stepsizes')
plt.legend()
plt.show()
```
