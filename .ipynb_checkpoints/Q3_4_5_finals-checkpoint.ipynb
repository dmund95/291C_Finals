```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
init_point = 1
num_iters = 100

def get_grad(point):
    return 4*point*point*point

def get_val(point):
    return point*point*point*point
    
def run_grad_descent(start, lr, iters):
    curr_point = start
    val_arr = []
    for i in range(iters):
        val_arr.append(get_val(curr_point))
        gred = get_grad(curr_point)
        curr_point = curr_point - lr*gred
    return val_arr
```

```python
val_arr1 = run_grad_descent(init_point, 0.1, num_iters)
val_arr2 = run_grad_descent(init_point, 0.01, num_iters)

plt.plot(val_arr1, label='lr=0.1')
plt.plot(val_arr2, label='lr=0.01')
plt.legend()
plt.show()
```

```python
def wolf1(start, step_size, c1):
    return (get_val(start - step_size*get_grad(start)) <= get_val(start) - c1*step_size*get_grad(start)*get_grad(start))

def wolf2(start, step_size, c2):
    return (get_grad(start - step_size*get_grad(start))*(-get_grad(start)) >= c2*get_grad(start)*(-get_grad(start)))
    
def get_step(start):
    init_step = 100000000000000
    c1 = 1e-4
    c2 = 0.9
    tau = 0.5
    while (1):
#         if (wolf1(start, init_step, c1) and wolf2(start, init_step, c2)):
        if (wolf1(start, init_step, c1)):
            return init_step
        init_step = tau*init_step

def run_wolf_descent(start, iters):
    curr_point = start
    val_arr = []
    for i in range(iters):
        val_arr.append(get_val(curr_point))
        gred = get_grad(curr_point)
        lr = get_step(curr_point)
        curr_point = curr_point - lr*gred
    return val_arr
```

```python
val_arr1 = run_grad_descent(init_point, 0.1, num_iters)
val_arr2 = run_grad_descent(init_point, 0.01, num_iters)
val_arr3 = run_wolf_descent(init_point, num_iters)

plt.plot(val_arr1, label='lr=0.1')
plt.plot(val_arr2, label='lr=0.01')
plt.plot(val_arr3, label='wolfe')
plt.legend()
plt.show()
```

```python
def run_newton_descent(start, lr, iters):
    curr_point = start
    val_arr = []
    for i in range(iters):
        val_arr.append(get_val(curr_point))
        curr_point = curr_point - lr*curr_point/3
    return val_arr
```

```python
val_arr1 = run_grad_descent(init_point, 0.1, num_iters)
val_arr2 = run_grad_descent(init_point, 0.01, num_iters)
val_arr3 = run_wolf_descent(init_point, num_iters)
val_arr4 = run_newton_descent(init_point, 1, num_iters)
val_arr5 = run_newton_descent(init_point, 0.1, num_iters)
val_arr6 = run_newton_descent(init_point, 0.01, num_iters)

plt.plot(val_arr1, label='steepest_descent_lr=0.1')
plt.plot(val_arr2, label='steepest_descent_lr=0.01')
plt.plot(val_arr3, label='wolfe')
plt.plot(val_arr4, label='newton_descent_lr=1')
plt.plot(val_arr5, label='newton_descent_lr=0.1')
plt.plot(val_arr6, label='newton_descent_lr=0.01')
plt.legend()
plt.show()
```
